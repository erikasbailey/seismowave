# Class Imbalance - Normal vs Extreme Sea Conditions - Percentiles

```python
results_quartiles = {}

grid_column_template = [f"Grid Cell {i+1}" for i in range(5)]

for station in station_list:
    mlvariables = pd.read_parquet(f"Processed Data/15.03.2025 Data/{station}.parquet")

    stats_cols = mlvariables.loc[:, mlvariables.columns.str.startswith("sea_")]
    grids_cols = mlvariables.loc[:, mlvariables.columns.str.startswith("(")]
    grids_cols.columns = ['Grid Cell 1', 'Grid Cell 2', 'Grid Cell 3', 'Grid Cell 4', 'Grid Cell 5']
    
    Y = pd.concat([stats_cols, grids_cols], axis=1)
    
    #compute percentiles for sea SWH
    percentile_values = Y.quantile(percentiles)
    
    #store results per station
    results_quartiles[station] = percentile_values

combined_quartiles = pd.concat(results_quartiles, names=["Station", "Percentile"])
```

```python
combined_quartiles
```

```python
variabletoplot = "sea_mean"

plot_data = combined_quartiles[variabletoplot].unstack(level=1)

plot_data.plot(kind='bar', figsize=(12, 6), cmap="GnBu", edgecolor='gray')
plt.grid(visible=True, alpha=0.5, color='gray', axis='y')
plt.title('Percentiles of mean SWH across stations')
plt.ylabel('Value')
plt.xlabel('Station')
plt.xticks(rotation=0)
plt.legend(title="Percentile")
plt.tight_layout()
plt.show()
```

# Error Analysis - MAE by SWH bin

## Intervals of 0.5m

```python
#swh ranges from 0 to 5.5 - evaluate in intervals of 0.5m
sea_mean_intervals = np.arange(0, 5.5, 0.5)

mae_by_station_and_range = {station: {interval: [] for interval in sea_mean_intervals} for station in station_list}

average_mae_by_station_and_range = {station: {interval: [] for interval in sea_mean_intervals} for station in station_list}
```

```python
for station, bestfoldnum in zip(station_list, kfoldbestfold):
    best_Ypred = pd.read_parquet(f"results/kfold_data/{station}/fold{bestfoldnum}_Ypred.parquet")
    best_YTest = pd.read_parquet(f"results/kfold_data/{station}/fold{bestfoldnum}_YTest.parquet")

    for interval in sea_mean_intervals:
        mask = (best_YTest.iloc[:, 5] >= interval) & (best_YTest.iloc[:, 5] < interval + 0.5)
        
        if mask.sum() > 0:
            mae = mean_absolute_error(best_YTest.loc[mask], best_Ypred.loc[mask])
            mae_by_station_and_range[station][interval].append(mae)

        if np.sum(mae_by_station_and_range[station][interval]) != 0:
            averagemaestationrange = np.average(mae_by_station_and_range[station][interval])
        else:
            averagemaestationrange = 0
        average_mae_by_station_and_range[station][interval].append(averagemaestationrange)
```

```python
df_mae = pd.DataFrame.from_dict(average_mae_by_station_and_range)
```

```python
df_mae
```

```python
data = []
intervals = df_mae.index.values
stations = df_mae.columns

#flatten the data
for interval in intervals:
    for station in stations:
        data.append((interval, station, df_mae.loc[interval, station][0]))

#convert to dataframe
mae_flat_df = pd.DataFrame(data, columns=['Interval', 'Station', 'MAE'])

fig, ax = plt.subplots(figsize=(12, 6))
station_count = len(stations)
interval_count = len(intervals)
bar_width = 0.8

cmap = cm.get_cmap('GnBu', station_count)
colors = [cmap(i) for i in range(station_count)]

x_positions = []
x_labels = []

for i, interval in enumerate(intervals):
    base = i * (station_count+1)
    for j, station in enumerate(stations):
        x = base + j
        mae = mae_flat_df[(mae_flat_df['Interval'] == interval) & (mae_flat_df['Station'] == station)]['MAE'].values[0]
        ax.bar(x, mae, width=bar_width, color=colors[j], edgecolor='gray', label=station if i == 0 else "")
        x_positions.append(x)
        x_labels.append(station)

ax.set_ylabel('Average MAE')
ax.set_title('Average MAE per Station grouped by Actual Mean SWH')
ax.set_xticks(x_positions)
ax.set_xticklabels(x_labels, rotation=90, fontsize=8)
for i, interval in enumerate(intervals):
    center_x = i * (station_count + 1) + (station_count - 1) / 2
    ax.text(center_x, -0.045 * mae_flat_df['MAE'].max(), f'{interval}-{interval+0.5}m',
            ha='center', va='top', fontsize=9, transform=ax.get_xaxis_transform())
ax.set_xlabel('Average MAE per Station grouped SWH', labelpad=30)
plt.grid(visible=True, alpha=0.5, color='gray', axis='y')
#ax.legend(title='Station', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
