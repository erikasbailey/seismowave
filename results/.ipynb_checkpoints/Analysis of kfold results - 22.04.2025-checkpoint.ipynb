```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
stations = ['AIO', 'CAVT', 'CSLB', 'HAGA', 'MSDA', 'MUCR', 'WDD']

means =     [0.402911, 0.850343, 0.642036, 0.663316, 0.838596, 0.798822, 0.877430]
std_devs =  [0.135529, 0.031919, 0.393528, 0.123203, 0.028458, 0.046775, 0.034077]
mins =      [0.271027, 0.823155, -0.05001, 0.455042, 0.795433, 0.754810, 0.820396]
maxs =      [0.577653, 0.895242, 0.889382, 0.754709, 0.865631, 0.866596, 0.912234]

x = np.arange(len(stations))

plt.figure(figsize=(10, 6))

plt.scatter(x, maxs, marker='^', color='green', label='Max')
plt.scatter(x, mins, marker='v', color='red', label='Min')

plt.errorbar(x, means, yerr=std_devs, fmt='o', capsize=5, label='Mean ± Std Dev', color='blue', alpha=0.7)

plt.xticks(x, stations)
plt.ylabel('Value')
plt.title('Mean ± Standard Deviation with Minimum and Maximum for R2 per Station')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()

plt.show()
```
