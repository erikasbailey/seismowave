# Query the data

```python
DS = copernicusmarine.open_dataset(dataset_id="med-hcmr-wav-rean-h")
DS
```

```python
#output directory
output_dir = r".\Sea data"

#date range
start_date = "2018-01-01"
end_date = "2021-12-31"
date_range = pd.date_range(start=start_date, end=end_date)

#for each day in the range
for date in date_range:
    #for each hour of the day
    for hour in range(24):
        #create a datetime object for the specific hour
        date_time = date + pd.Timedelta(hours=hour)

        #extract the data for the specific datetime
        subset = DS[['VHM0']].sel(time=date_time, method='nearest')

        #rename the variables
        #subset = subset.rename({'latitude': 'lat', 'longitude': 'lon', 'VHM0': 'Y', 'time': 'time'})
        
        #generate filenames with date and time
        date_str = date_time.strftime("%Y%m%d") #format: YYYYMMDD
        time_str = f"{hour:02d}00" #format: HH00
        
        nc_filename = os.path.join(output_dir, f"{time_str}_{date_str}.nc")
        #mat_filename = os.path.join(output_dir, f"{time_str}_{date_str}.mat")
        #.mat not required
        
        #save to NetCDF
        subset.to_netcdf(nc_filename)
        
        #Convert to .mat and save
        
        #data_dict = {'VHM0': subset['VHM0'].values}
        #savemat(mat_filename, data_dict)

        #print(f"Saved files for {date_time}: {nc_filename} and {mat_filename}")
```
